```python
usda_trees = usda_trees[usda_trees['MatureHeight'] >= 4.0]
usda_trees['MatureHeight'].tolist()
```

```python
df1 = usda_trees.copy()


df1['State'] = df1['State'].str.split('> ').str[1]
df1['State'] = df1['State'].str.split(' <strong').str[0]


import ast
df1 = df1.dropna( subset=['State'], axis=0)
df1['State'] = df1['State'].apply( lambda x: x.replace("(","").replace(")","").split(", ") )


df2 = df1[['CommonName',"State"]].copy()

a = []
for i,r in df2.iterrows():

    treename = r['CommonName']

    state_list = r['State']

    for state in state_list:
        a.append( [treename,state] )


named_df = pd.DataFrame(
    data = a,
    columns= ['CommonName','State']
)

named_df


```

```python
named_df["State"] = named_df["State"].str.replace(',','')
named_df
```
